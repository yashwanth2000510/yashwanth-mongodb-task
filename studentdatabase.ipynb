# 2)      Find students who scored below average in the exam and pass mark is 40%?


```python
# below average is between 40% to 60% 
# first class is between 60% to  80%
# distinction is above 80%
```

```python
for i in information.aggregate([{"$project":{"name":1,"total_marks":{"$max":"$scores.score"}}},
                                {"$match":{"$or":[{"total_marks":{"$gte":40,"$lte":60}}]}},
                               {"$sort":{"total_marks":1}}]):
    print(i)
```

# 3)      Find students who scored below pass mark and assigned them as fail, and above pass mark as pass in all the categories.


### students who failed in all the categories

```python
A= information.find({"scores.0.score": {"$lt": 40}, "scores.1.score": {"$lt": 40}, "scores.2.score": {"$lt": 40}}, {"_id": -1, "name":-1})
for i in A:
    print(i)
```

### students who have passed in all the categories

```python
A= information.find({"scores.0.score": {"$gte": 40}, "scores.1.score": {"$gte": 40}, "scores.2.score": {"$gte": 40}}, {"_id": 1, "name":1})
for i in A:
    print(i)
```

# 4)       Find the total and average of the exam, quiz and homework and store them in a separate collection.

```python
a =  information.aggregate([{"$project":{"total_marks":{"$sum":"$scores.score"},
                                         "average":{"$avg":"$scores.score"}}}])
  
```

```python
for i in a:
    mydb.total_avg.insert_one(i)
```

```python
for i in mydb.total_avg.find():
    print(i)
```
